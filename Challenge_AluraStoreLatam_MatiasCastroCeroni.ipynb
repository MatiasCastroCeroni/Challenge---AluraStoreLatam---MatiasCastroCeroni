```python
# Resumen general de facturación por tienda
print("\n------------------------------------------")
print("=== Resumen de Facturación por Tienda ===")
print("------------------------------------------")
print(f"Tienda 1: ${facturacion_tienda1:,.2f}")
print(f"Tienda 2: ${facturacion_tienda2:,.2f}")
print(f"Tienda 3: ${facturacion_tienda3:,.2f}")
print(f"Tienda 4: ${facturacion_tienda4:,.2f}")
print("------------------------------------------\n")
```

```python
#Grafico análisis facturación por Tienda
# Crear una lista o diccionario con los resultados
tiendas = ['Tienda 1', 'Tienda 2', 'Tienda 3', 'Tienda 4']
facturacion = [facturacion_tienda1, facturacion_tienda2, facturacion_tienda3, facturacion_tienda4]

```

```python
#Grafico análisis facturación por Tienda
import matplotlib.pyplot as plt

# Configuración del gráfico
plt.figure(figsize=(10,6))
plt.bar(tiendas, facturacion, color='green')

# Títulos y etiquetas
plt.title('Facturación Total por Tienda', fontsize=16, fontweight='bold')
plt.xlabel('Tiendas', fontsize=12)
plt.ylabel('Facturación', fontsize=12)

# Mostrar los valores arriba de cada barra
for i, valor in enumerate(facturacion):
    plt.text(i, valor + (valor*0.02), f"${valor:,.0f}", ha='center', fontsize=10, color='black')

# Mostrar el gráfico
plt.tight_layout()
plt.show()
```

# 2. Ventas por categoría

```python
#Categoria más vendida en tienda 3
ventas_por_categoria_t3 = tienda3.groupby('Categoría del Producto').size().sort_values(ascending=False)
categoria_top_t3 = ventas_por_categoria_t3.idxmax()
ventas_top_t3 = ventas_por_categoria_t3.max()
print(categoria_top_t3, ventas_top_t3)
```

```python
#Gráfico análisis productos más y menos vendidos por tienda
import matplotlib.pyplot as plt
import numpy as np

# Datos
tiendas = ['Tienda 1', 'Tienda 2', 'Tienda 3', 'Tienda 4']
mas_vendidos = [cantidad_mas_vendido_t1, cantidad_mas_vendido_t2, cantidad_mas_vendido_t3, cantidad_mas_vendido_t4]
menos_vendidos = [cantidad_menos_vendido_t1, cantidad_menos_vendido_t2, cantidad_menos_vendido_t3, cantidad_menos_vendido_t4]

# Posición de las barras
x = np.arange(len(tiendas))
ancho = 0.35

# Configuración del gráfico
plt.figure(figsize=(10,6))
barras1 = plt.bar(x - ancho/2, mas_vendidos, ancho, label='Más vendidos', color='#2ECC71', edgecolor='black')
barras2 = plt.bar(x + ancho/2, menos_vendidos, ancho, label='Menos vendidos', color='#E74C3C', edgecolor='black')

# Títulos y etiquetas
plt.title('Productos Más y Menos Vendidos por Tienda', fontsize=16, fontweight='bold', pad=25)
plt.xlabel('Tiendas', fontsize=12)
plt.ylabel('Cantidad de Ventas', fontsize=12)
plt.xticks(x, tiendas)
plt.legend()

# Mostrar valores encima de las barras
for barras in [barras1, barras2]:
    for barra in barras:
        yval = barra.get_height()
        plt.text(barra.get_x() + barra.get_width()/2, yval + 0.5, f"{int(yval)}",
                 ha='center', va='bottom', fontsize=9, color='black')

# Estilo general
plt.grid(axis='y', linestyle='--', alpha=0.6)
plt.tight_layout()
plt.show()
```
